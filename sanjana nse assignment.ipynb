```python

# Student Record Manager
# Python Mini Project

students = [{"ID": 1, "Name": "Alice", "Age": 20, "Department": "CSE"},
{"ID": 2, "Name": "Asha", "Age": 21, "Department": "HR"},
{"ID": 3, "Name": "Usha", "Age": 20, "Department": "SCM"},
{"ID": 4, "Name": "Akash", "Age": 22, "Department": "FINANCE"},
{"ID": 5, "Name": "Anu", "Age": 25, "Department": "ANALYTICS"},
{"ID": 6, "Name": "Pooja", "Age": 21, "Department": "CIVIL"},
{"ID": 7, "Name": "Kiran", "Age": 22, "Department": "ME"},
{"ID": 8, "Name": "Nisha", "Age": 20, "Department": "ECE"},
{"ID": 9, "Name": "Arjun", "Age": 21, "Department": "CSE"},
{"ID": 10, "Name": "Divya", "Age": 22, "Department": "EEE"},
{"ID": 11, "Name": "Rohan", "Age": 23, "Department": "CSE"},
{"ID": 12, "Name": "Megha", "Age": 20, "Department": "CIVIL"},
{"ID": 13, "Name": "Suresh", "Age": 21, "Department": "ME"},
{"ID": 14, "Name": "Kavya", "Age": 22, "Department": "ECE"},
{"ID": 15, "Name": "Aditya", "Age": 20, "Department": "CSE"},
{"ID": 16, "Name": "Bhavya", "Age": 21, "Department": "EEE"},
{"ID": 17, "Name": "Manoj", "Age": 23, "Department": "ME"},
{"ID": 18, "Name": "Shruti", "Age": 20, "Department": "CIVIL"},
{"ID": 19, "Name": "Naveen", "Age": 22, "Department": "ECE"},
{"ID": 20, "Name": "Isha", "Age": 21, "Department": "CSE"},
{"ID": 21, "Name": "Prakash", "Age": 23, "Department": "EEE"},
{"ID": 22, "Name": "Tanvi", "Age": 20, "Department": "CIVIL"},
{"ID": 23, "Name": "Sanjay", "Age": 22, "Department": "ME"},
{"ID": 24, "Name": "Ritu", "Age": 21, "Department": "ECE"},
{"ID": 25, "Name": "Harsha", "Age": 20, "Department": "CSE"},
{"ID": 26, "Name": "Neha", "Age": 22, "Department": "EEE"},
{"ID": 27, "Name": "Amit", "Age": 23, "Department": "ME"},
{"ID": 28, "Name": "Lakshmi", "Age": 21, "Department": "CIVIL"},
{"ID": 29, "Name": "Varun", "Age": 22, "Department": "ECE"},
{"ID": 30, "Name": "Riya", "Age": 20, "Department": "CSE"}
]

    



# List to store student records

# Function to add a student
def add_student():
    try:
        student_id = int(input("Enter Student ID: "))
        name = input("Enter Student Name: ").strip().title()
        age = int(input("Enter Age: "))
        department = input("Enter Department: ").strip().upper()

        student = {
            "ID": student_id,
            "Name": name,
            "Age": age,
            "Department": department
        }

        students.append(student)
        print("Student record added successfully!")

    except ValueError:
        print("Invalid input! ID and Age must be numbers.")

# Function to view all students
def view_students():
    if len(students) == 0:
        print("No student records found.")
    else:
        print("\n--- Student Records ---")
        for student in students:
            print("ID:", student.get("ID"))
            print("Name:", student.get("Name"))
            print("Age:", student.get("Age"))
            print("Department:", student.get("Department"))
            print("----------------------")

# Function to search student by department
def search_student():
    search_dept = input("Enter department to search: ").strip().upper()
    found = False
    for student in students:
        if student.get("Department") == search_dept:
            print(student)
            found = True
    if not found:
        print("No student found in this department.")

# Function to save student records to file
def save_to_file():
    with open("students.txt", "w") as file:
        for student in students:
            file.write(str(student) + "\n")
    print("Student records saved to file.")


while True:
    print("\n===== Student Record Manager Menu =====")
    print("1. Add Student")
    print("2. View Students")
    print("3. Search Student by Department")
    print("4. Save & Exit")

    choice = input("Enter your choice (1-4): ").strip()

    if choice == "1":
            add_student()
    elif choice == "2":
            view_students()
    elif choice == "3":
            search_student()
    elif choice == "4":
        save_to_file()
        print("Exiting program. Thank you!")
        break
    else:
        print("Invalid choice. Please select 1 to 4.")

```
